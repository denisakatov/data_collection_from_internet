```python
source = 'opendata.csv'
df = pd.read_csv(source, encoding = 'windows-1251', parse_dates=['date'], index_col='date')
date_filter = df.loc[date_start:date_finish]
result = date_filter[(date_filter['value']) & (date_filter['name'] == 'Средняя сумма заявки на потребительский кредит')&(date_filter['region'] == region_input)]['value']
print(result)
```

```python
# data_frame.loc['2014-01-01':'2014-12-01']
result.plot()
plt.title(f'{type_input} в \n {region_input} за период с {date_start}  по {date_finish} ')
plt.xlabel("Дата")
plt.ylabel("Количество заказов")
plt.legend()
plt.rcParams['figure.figsize'] = (16, 8)
```
